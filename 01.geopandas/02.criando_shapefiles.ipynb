#  Criando Shapefiles

## Obtendo arquivos

https://geoftp.ibge.gov.br/organizacao_do_territorio/malhas_territoriais/malhas_municipais/municipio_2019/Brasil/BR/br_unidades_da_federacao.zip


```python
%matplotlib inline
import geopandas as gpd
```

```python
geobr = gpd.read_file('dados/originais/br_unidades_da_federacao/BR_UF_2019.shp')
```

```python
geobr
```

```python
geobr.plot(color='white', edgecolor='black', figsize=(15,8))
```

### Modificando 

```python
ufgo = geobr.query('SIGLA_UF == "GO"')
ufgo
```

```python
ufgo.plot(color='orange', edgecolor='black', figsize=(15,8))
```

### Salvando

```python
import os

dir_ = 'dados/processados/mapas/GO-MUNIC' 
if not os.path.exists(dir_):
    os.makedirs(dir_)
```

```python
ufgo.to_file(dir_ + '/GO-MUNIC.shp')
```

```python
go_munic = gpd.read_file(dir_ + '/GO-MUNIC.shp')
```

```python
go_munic.plot(color='orange', edgecolor='black', figsize=(15,8))
```
